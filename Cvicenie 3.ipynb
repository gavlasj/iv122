#### A) Knihovna pro želví grafiku

turtle.py:

```python
import math
import svglib

class turtle:
    def __init__(self):
        self.cx = 100
        self.cy = 100
        self.angle = 0
        self.lines = []

    def forward(self,length):
        newx = math.cos(math.radians(self.angle))*length + self.cx
        newy = math.sin(math.radians(self.angle))*length + self.cy
        self.lines.append((self.cx, self.cy,newx,newy))
        
        self.cx = newx
        self.cy = newy

    def back(self,length):
        forward(self,-length)
    def right(self,angle):
        self.angle += angle
    def left(self,angle):
        self.angle += (360-angle)

    def save(self,fileName):
        file = open (fileName + ".svg", "w")
        file.write(svglib.header())

        for x1,y1,x2,y2 in self.lines:
            file.write(svglib.line(x1,y1,x2,y2))
        
        file.write(svglib.footer())
        file.close()
```

#### B) Vykreslování relativně vs. absolutně

```python
from IPython.display import SVG, display
import turtle

tut = turtle.turtle()

def pentagram(n):
    for i in range(5):
        tut.forward(n)
        tut.left(360/5)
    tut.left((180-360/5)/3)
    for i in range(5):
        tut.forward(((1 + math.sqrt(5)) / 2)*n)
        tut.left(720/5)
    
    tut.save("pentagram")
    display(SVG(filename = "pentagram.svg"))
    return

pentagram(50)
```

```python
from IPython.display import SVG, display
import turtle

tut = turtle.turtle()

def spiral(leng,steps):
    for i in range(steps):
        for i in range(4):
            tut.forward(leng)
            tut.left(90)
        tut.forward(leng/4)
        tut.left(18.5)
        leng = math.sqrt((leng/4)**2+(leng-(leng/4))**2)
        
    tut.save("squares")
    display(SVG(filename = "squares.svg"))
    return

spiral(80,7)
```

```python
from IPython.display import SVG, display
import turtle

tut = turtle.turtle()

def flower(len):
    for i in range(12):
        for j in range(6):
            tut.forward(len)
            tut.left(30)
        tut.forward(len)
        for j in range(6):
            tut.forward(len)
            tut.left(30)
        tut.forward(len)
        tut.left(30)
            
    tut.save("diamond")
    display(SVG(filename = "diamond.svg"))
    return

flower(12)
```

#### C) Želví grafika a fraktály

```python
from IPython.display import SVG, display
import turtle

tut = turtle.turtle()

def ker(length,level):
    if level == 1:
        tut.forward(length)
        
        tut.left(180)
        tut.forward(length)
        tut.left(180)
        #tut.back(length)
    else:
        tut.forward(length)
        tut.left(45)
        ker(length/2,level-1)

        tut.right(90)
        ker(length/2,level-1)
        tut.left(45)
        
        tut.left(180)
        tut.forward(length)
        tut.left(180)
        #tut.back(length)
        
    return

ker(60,5)
tut.save("ker")  
display(SVG(filename = "ker.svg"))  
```

```python
from IPython.display import SVG, display
import turtle

tut = turtle.turtle()

#pomocna
def koch(leng,n):
    if n==1:
        tut.forward(leng)
    else:
        koch(leng/3,n-1)
        tut.right(60)
        koch(leng/3,n-1)
        tut.left(120)
        koch(leng/3,n-1)
        tut.right(60)
        koch(leng/3,n-1)
    return

def koch_flake(leng,n):
    for i in range(3):
        koch(leng,n)
        tut.left(120)
        
    tut.save("koch")
    display(SVG(filename = "koch.svg"))
    
    return 

koch_flake(110,5)
```

```python
from IPython.display import SVG, display
import turtle

tut = turtle.turtle()

def triangle(leng,step):
    if step == 1:            
        for i in range(3):
            tut.forward(leng)
            tut.left(120)
    else:
        for i in range(3):
            triangle(leng/2,step-1)
            tut.forward(leng)
            tut.left(120)
    return

triangle(110,5)
tut.save("sierp")
display(SVG(filename = "sierp.svg"))
```
